#Exercise3 - PCP from Starch

##STEP 1: Load & Prepare Data

```python
# Convert to grayscale first
gray_img = Image.open('/content/drive/MyDrive/AI and Machine Learning/image/w3.png').convert('L')
X = np.array(gray_img, dtype=float)

plt.imshow(X, cmap='gray')
plt.title("Original Grayscale")
plt.axis('off')
plt.show()
```

###Standardize (Center the Data)

```python
mean = np.mean(X, axis=0)
X_centered = X - mean
```

###Covariance Matrix

```python
cov_matrix = np.cov(X_centered, rowvar=False)
```

##STEP2: Eigen Decomposition

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
```

###Sort Eigenvalues

```python
idx = np.argsort(eigenvalues)[::-1]

eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]
```

###Cumulative Explained Variance

```python
explained_variance = eigenvalues / np.sum(eigenvalues)
cumulative_variance = np.cumsum(explained_variance)

plt.plot(cumulative_variance)
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.show()
```

##STEP 3: Reconstruction

```python
def reconstruct(k):
    P = eigenvectors[:, :k]
    X_reduced = np.dot(X_centered, P)
    X_reconstructed = np.dot(X_reduced, P.T) + mean
    return X_reconstructed
```

```python
k_values = [5, 25, 50, 100]

plt.figure(figsize=(15,10))

for i, k in enumerate(k_values):
    img_recon = reconstruct(k)

    plt.subplot(2,2,i+1)
    plt.imshow(img_recon.real, cmap='gray')
    plt.title(f"{k} Components")
    plt.axis('off')

plt.show()
```
